```python
# Función para convertir datos binarios IDX en arrays de Numpy:
def read_idx(filename):
    with open(filename, 'rb') as f: # 'rb' = leer en modo binario
        #Para leer la cabecera del archivo IDX
        zero, data_type, dims = struct.unpack('>HBB', f.read(4))
        #Para leer dimensiones:
        shape = tuple(struct.unpack('>I', f.read(4))[0] for d in range(dims))
        #leer los datos y convertir a array.
        data = np.frombuffer(f.read(), dtype=np.uint8)
        return data.reshape(shape)
```
